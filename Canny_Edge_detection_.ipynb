```python
import numpy as np
import matplotlib.pyplot as plt
from cv2 import imread
from skimage.color import rgb2gray
from skimage import transform
from skimage.exposure import equalize_hist
import os
from skimage.color import rgb2gray
from skimage import transform
from skimage.exposure import equalize_hist
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```python

def preprocess(img):
    img = transform.resize(img, (128, 128))
    img = rgb2gray(img)
    img = equalize_hist(img)
    return img
```

```python
img = imread('D:\Desktop\CV Project\data\lionel_messi\lionel_messi1.png')
```

```python
img = preprocess(img)
```

```python
img.shape
```

```python
def convert_to_image(image):
    image = image * 255
    image = image.astype(np.uint8)
    return image
```

```python
#convert nparray to img
img = convert_to_image(img)
```

```python
edges = cv.Canny(img, 200, 100)
```

```python

plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python


def asImage(arr):
    arr = arr.astype(np.uint8)
    val =  cv2.cvtColor(arr, cv2.COLOR_GRAY2BGR)
    return val
```

```python
def canny_transform():
    X = []
    Y = []
    # print(os.getcwd())
    for i in os.listdir(os.getcwd()+'/data'):
        for j in os.listdir(os.getcwd()+'/data/'+i):
            image = imread(os.getcwd()+'/data/'+i+'/'+j)
            image = preprocess(image)
            image = image * 255
            image = image.astype(np.uint8)
            edges = cv2.Canny(image, 200, 100)
            X.append(asImage(edges))
            Y.append(i)
    X = np.asarray(X)
    Y = np.asarray(Y)
    # print(X)
    return X, Y
```
